## Question One – Magic 8 Ball

This program simulates a Magic 8 Ball by accepting a user's yes/no question
and randomly generating a response from predefined categories.


```python
import random

affirmative = [
    "It is certain",
    "Yes definitely",
    "Most likely"
]

non_committal = [
    "Reply hazy, try again",
    "Ask again later",
    "Cannot predict now"
]

negative = [
    "Don't count on it",
    "My reply is no",
    "Very doubtful"
]

question = input("Ask a yes/no question: ")
print("\nMagic 8 Ball says:")
print(random.choice(affirmative + non_committal + negative))
```

**Note:**
This cell requires user input.
Type any yes/no question and press Enter to receive a Magic 8 Ball response.


**Output Explanation:**
The program randomly selected a non-committal Magic 8 Ball response.
Each execution may produce a different result due to random selection.


## Fortune Cookie

This program simulates a fortune cookie by displaying a random fortune
and four lucky numbers between 1 and 99.


```python
fortunes = [
    "You will have a pleasant surprise.",
    "A new opportunity is coming your way.",
    "Good news will arrive soon.",
    "Your hard work will pay off.",
    "Happiness begins with you.",
    "A fresh start is ahead."
]

lucky_numbers = [random.randint(1, 99) for _ in range(4)]

print("Fortune:")
print(random.choice(fortunes))
print("Lucky Numbers:", lucky_numbers)
```

**Output Explanation:**
The program randomly selected a fortune message and generated four
lucky numbers between 1 and 99 using Python's random module.
Each execution produces different results.


## Question Two – Fizz Buzz

This program prints numbers from 1 to 100.
- Multiples of 3 print "Fizz"
- Multiples of 5 print "Buzz"
- Multiples of both print "FizzBuzz"


```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

**Output Explanation:**
The program correctly replaces multiples of 3 with "Fizz",
multiples of 5 with "Buzz", and multiples of both with "FizzBuzz"
while counting from 1 to 100.
