```python
# Magic 8 Ball Program
# The user enters a yes/no question
# The program prints a random response

import random

affirmative = [
    "It is certain",
    "Yes definitely",
    "Most likely"
]

non_committal = [
    "Reply hazy, try again",
    "Ask again later",
    "Cannot predict now"
]

negative = [
    "Don't count on it",
    "My reply is no",
    "Very doubtful"
]

question = input("Ask a yes/no question: ")
print("\nMagic 8 Ball says:")
print(random.choice(affirmative + non_committal + negative))
```

```python
# Fortune Cookie Program
# Prints a random fortune and four lucky numbers between 1 and 99

fortunes = [
    "You will have a pleasant surprise.",
    "A new opportunity is coming your way.",
    "Good news will arrive soon.",
    "Your hard work will pay off.",
    "Happiness begins with you.",
    "A fresh start is ahead."
]

lucky_numbers = [random.randint(1, 99) for _ in range(4)]

print("Fortune:")
print(random.choice(fortunes))
print("Lucky Numbers:", lucky_numbers)
```

```python
# Fizz Buzz Program
# Prints numbers from 1 to 100
# Replaces multiples of 3 with "Fizz"
# Replaces multiples of 5 with "Buzz"
# Replaces multiples of both with "FizzBuzz"

for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
